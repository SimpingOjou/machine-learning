# Saliency detector using OpenCV

```python
# import the necessary packages
import cv2
import matplotlib.pyplot as plt
import numpy as np
import argparse

# define the image path directly
image_path = "./img/mousey.png"

# load the input image
image = cv2.imread(image_path)
```

## Spectral saliency

```python
# initialize OpenCV's static saliency spectral residual detector and compute the saliency map
saliency = cv2.saliency.StaticSaliencySpectralResidual_create()
(success, saliencyMap) = saliency.computeSaliency(image) # [0, 1]
saliencyMap = (saliencyMap * 255).astype("uint8") # scale it to the range [0, 255]

# Display the images using matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Saliency Map")
plt.imshow(saliencyMap, cmap='gray')
plt.axis('off')

plt.show()
```

## Fine grated saliency

```python
saliency = cv2.saliency.StaticSaliencyFineGrained_create()
(success, saliencyMap) = saliency.computeSaliency(image)

threshMap = cv2.threshold(saliencyMap.astype("uint8"), 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

# Display the images using matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Saliency Map")
plt.imshow(saliencyMap, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Threshold Map")
plt.imshow(threshMap, cmap='gray')
plt.axis('off')

plt.show()
```
